```python
import os
import time
import requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


import lightgbm as lgb


import warnings
warnings.filterwarnings('ignore')
```

```python
class Config:
  _ALPHA_VANTAGE_API_KEY = 'FK2PGNI330AEKBQL'
  _TS_FEAT = {
                'open':'1. open', 
                'high':'2. high', 
                'low':'3. low', 
                'close':'4. close', 
                'volume':'5. volume'
  }
  _TICKER_DIR = '/content/drive/MyDrive/AV Portfolio ML/stock-ticker.xlsx'
  _SEED = 42
```

```python
def seed_everything(seed):
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)

seed_everything(Config._SEED)
```

```python
ticker_sp500 = pd.read_excel(Config._TICKER_DIR)
ticker_sp500.head()
```

```python
ticker_list_sp500 = list(ticker_sp500['TICKER'].values)
```

```python
print(ticker_list_sp500)
```

```python
for i, t in enumerate(ticker_list_sp500):
  if t == 'TSCO':
    print(i)
```

```python
for i, ticker in enumerate(ticker_list_sp500):
  if i % 5 == 0 and i != 0:
    print('[Info] API calls per minute reached - program sleeps for 1 minute')
    time.sleep(60)
  print(f'[Info] Fetching {ticker}')
  url = f'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol={ticker}&outputsize=full&apikey={Config._ALPHA_VANTAGE_API_KEY}'
  r = requests.get(url)
  data = r.json()
  
  try:
    ts_dict = {}
    ts_dates = np.array(list(data[list(data.keys())[1]])[::-1])
    ts_dict['date'] = ts_dates
    for key, val in Config._TS_FEAT.items(): ts_dict[key] = np.array([data[list(data.keys())[1]][key][val] for key in ts_dates])

    ts_df = pd.DataFrame.from_dict(ts_dict)
    ts_df['ticker'] = ticker
    ts_df.to_excel(f'/content/drive/MyDrive/AV Portfolio ML/S&P 500 Stock Data/{ticker}.xlsx')
  except:
    print(f'[Warning] Ticker {ticker} not fetched')
```
